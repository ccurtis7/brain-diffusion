```python
def return_average(n1, n2, n3):

    f1 = np.zeros((frames, 3))

    for i in range(0, frames):
        f1[i, 0] = n1[i]
        f1[i, 1] = n2[i]
        f1[i, 2] = n3[i]

    answer = np.mean(f1, axis=1)
    return answer
```
