```python
def compute_geomean(df, ColumnName):
    g = [0]
    for i in range(2, len(df)+1):
        a = df[ColumnName].ix[i]
        b = list(a[[0,1,2]])
        d= []
        for j in b:
            c = float(j)
            d.append(c)
        g.append(scipy.stats.gmean(d))
    df[ColumnName+' geo'] = g
```

```python
compute_geomean(msd,'PEG-PLGA4A UP')
msd
```

```python
for title in columns:
    compute_geomean(msd, title)
```

```python
def compute_plot_Deff(ParticleChemistry,tmin,tmax):
    temp1_msd = msd[msd.index >= tmin]
    temp2_msd = temp1_msd[temp1_msd.index <= tmax]
    temp2_msd[[ParticleChemistry, u'PEG-PLGA4A UP geo',      u'PEG-PLGA15k UP geo',
            u'PEG-PLGA4A P80 geo',     u'PEG-PLGA15k P80 geo',
            u'PEG-PLGA4A F68 geo',     u'PEG-PLGA15k F68 geo',
          u'PEG-PLGA4A 5%CHA geo',   u'PEG-PLGA15k 5%CHA geo',
          u'PEG-PLGA15k 2CHA geo',  u'PEG-PLGA15k 0.5CHA geo',
          u'PLGA 4A UP-Exp 1 geo', u'PEG-PLGA4A UP-Exp 1 geo',
                u'PLGA58k UP geo',          u'PLGA15k UP geo',
                u'PLGA4A P80 geo',         u'PLGA15k P80 geo',
                u'PLGA4A F68 geo',         u'PLGA15k F68 geo',
              u'PLGA4A 5%CHA geo',       u'PLGA15k 5%CHA geo',
              u'PLGA15k 2CHA geo',      u'PLGA15k 0.5CHA geo']].plot(legend=False)
    Deff = scipy.stats.gmean(temp2_msd[ParticleChemistry])
    print Deff
```
